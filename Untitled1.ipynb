```python
import pandas as pd

# Replace with your actual GitHub repo raw link once you upload sales_data.csv
url = "https://raw.githubusercontent.com/Samuelibitola/Sales-Insights-Dashboard/main/sales_data.csv"

# Load CSV
df = pd.read_csv(url)
df.head()
```

```python
# Basic stats
df.describe()

# Total sales per region
df["Sales"] = df["Quantity"] * df["Price"]
region_sales = df.groupby("Region")["Sales"].sum()
region_sales
```

```python
# Sales Insights Dashboard - Demo
import pandas as pd
import matplotlib.pyplot as plt

# Load data - Data is already loaded in a previous cell

# Basic info
print("First 5 rows:")
print(df.head())

# Total revenue per product
df["Revenue"] = df["Quantity"] * df["Price"]
product_sales = df.groupby("Product")["Revenue"].sum()

# Visualization
plt.bar(product_sales.index, product_sales.values)
plt.title("Revenue by Product")
plt.ylabel("Revenue ($)")
plt.xlabel("Product")
plt.show()

# Regional insights
region_sales = df.groupby("Region")["Revenue"].sum()
print("\nRevenue by Region:")
print(region_sales)
```
